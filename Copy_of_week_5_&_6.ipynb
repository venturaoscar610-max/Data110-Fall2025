```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
titanic_df = sns.load_dataset('titanic')
print(titanic_df.head())
print(titanic_df.shape)
print(titanic_df.describe())
```

These columns matter to see how many people survied. What was the largest age demographic on the boat. Find the median age for each Class on the boat. I would use the sex columns to see if there are more females or males on board the ship. How much the boat made in tickets sales per class.

```python
plt.figure(figsize=(10, 6))
plt.hist(titanic_df['fare'], bins=30, color='steelblue', edgecolor='white')
plt.title('Fare Distribution: Histogram', fontsize=12, fontweight='bold')
plt.xlabel('Fare (pounds)', fontsize=11)
plt.ylabel('Number of Passengers', fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

Most of the passangers are paying 0-20 dollars. These are the cheapest. With a few paying in the one hundreds to 260 range in pounds. With one outlier of 500 pounds, which would've been super luxury compared to the pthers. Majority of the boat carried  poor passangers with a few super rich.

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(titanic_df['fare'], fill=True, color='coral', linewidth=2.5)
plt.title('Fare Distribution: KDE', fontsize=12, fontweight='bold')
plt.xlabel('Fare (pounds)', fontsize=11)
plt.ylabel('Density', fontsize=11)
plt.xlim(0, titanic_df['fare'].max())
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(11, 6))
sns.kdeplot(
titanic_df[titanic_df['survived'] == 0]['fare'],
label='Did Not Survive (0)',
color='darkred',
linewidth=2.5,
fill=True,
alpha=0.4,
bw_adjust=0.7
)
```

```python
plt.figure(figsize=(11, 6))
sns.kdeplot(
titanic_df[titanic_df['survived'] == 1]['fare'],
label='Survived (1)',
color='darkgreen',
linewidth=2.5,
fill=True,
alpha=0.4,
bw_adjust=0.7
)
```

For the peaks information we are getting the same display in values. The same outliers can be seen from this data as well.

```python
plt.figure(figsize=(11, 6))

sns.kdeplot(
titanic_df[titanic_df['survived'] == 0]['fare'],
label='Did Not Survive (0)',
color='darkred',
linewidth=2.5,
fill=True,
alpha=0.4,
bw_adjust=0.7
)

sns.kdeplot(
titanic_df[titanic_df['survived'] == 1]['fare'],
label='Survived (1)',
color='darkgreen',
linewidth=2.5,
fill=True,
alpha=0.4,
bw_adjust=0.7
)

plt.title('Fare Distribution by Survival Status', fontsize=13, fontweight='bold')
plt.xlabel('Fare (pounds)', fontsize=11)
plt.ylabel('Density', fontsize=11)
plt.legend(fontsize=10)
plt.xlim(0, titanic_df['fare'].max())
plt.grid(axis='y', alpha=0.3)
plt.show()
```

Survirours had better accesss to life boats. The Poor passengers probaly got locked in their class level and didn't have access to life boats. The lower class came second to higher class passengers which is why we see a spike in deaths for cheaper class sales.
